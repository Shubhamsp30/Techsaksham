```python
def plot_histograms(df, numerical_cols):
    for col in numerical_cols:
        plt.figure(figsize=(6, 4))
        sns.histplot(df[col], bins=20, kde=True, color="blue")
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()
numerical_cols = ['Total', 'Unit price', 'Rating']
plot_histograms(df, numerical_cols)
```

```python
def histogram_matplotlib(vari, bins=10):  
    var = df[vari]
    varValue=var.value_counts()
    plt.figure(figsize=(9,3))
    plt.hist(var, bins=bins)  
    plt.xlabel(vari)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {vari}')
    plt.show()
    print(f'{vari} (Histogram Data):\nÂ {varValue}')
```

```python
columns_plot=['Total','Unit price','Rating']
for i in columns_plot:
   histogram_matplotlib(i)
```

```python
def plot_boxplots(df, numerical_cols):
    plt.figure(figsize=(12, 10))
    for col in numerical_cols:
        plt.figure(figsize=(5, 4))
        sns.boxplot(y=df[col], color="cyan")
        plt.title(f'Boxplot of {col}')
        plt.ylabel(col)
        plt.show()
numerical_cols = ['Total','Quantity','Rating','gross income']
plot_boxplots(df, numerical_cols)
```

```python
def plot_bar_charts(df, categorical_cols):
    for col in categorical_cols:
        plt.figure(figsize=(8,5))
        sns.countplot(x=df[col], palette="coolwarm")
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=30)
        plt.show()
categorical_cols = ['Customer type','Payment','Product line']
plot_bar_charts(df, categorical_cols)
```

```python
def plot_pie_chart(df, column):
    data = df[column].value_counts() 
    labels = data.index 
    plt.figure(figsize=(6, 6))
    plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'orange'])
    plt.title(f'Distribution of {column}')
    plt.show()
plot_pie_chart(df,'Quantity') 
```

```python
def plot_scatter_plot(df, x_col, y_col):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[x_col], y=df[y_col], color="blue", alpha=0.6, edgecolor='black')
    plt.title(f'Scatter Plot of {x_col} vs {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.show()
plot_scatter_plot(df,'Unit price','gross income') 
```

```python
def plot_heatmap(df):
    numerical_df = df.select_dtypes(include=['number'])  
    plt.figure(figsize=(8, 6))
    correlation_matrix = numerical_df.corr()
    sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",linewidths=0.5,fmt=".2f")
    plt.title("Heatmap of Numerical Features")
    plt.show()
plot_heatmap(df)
```
